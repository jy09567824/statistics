# Probablity Distributions

```python
import numpy as np
from scipy import stats
import math
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

## Binomial Distribution

```python
# 投擲硬幣 500 次, 出現的正面的次數, 實驗 1000 次
n, p = 500, 0.5
arr_bi = np.random.binomial(n, p, 1000)

# WLLN: 重複實驗機率多次, 多次實驗的平均結果近似於期望值
print('E(x):', n * p)
print('Mean of all trials:', arr_bi.mean())

sns.histplot(data=arr_bi, kde=True)
```

```python
# n=10, p=0.05
n, p = 10, 0.05
arr_bi = np.random.binomial(n=10, p=0.05, size=10000)

# P(X >= 2)
len(arr_bi[arr_bi >= 2]) / 10000
```

## Normal Distribution

```python
# 代入標準常態的 pdf 公式計算
x = np.arange(start=-4, stop=4, step=0.1)
y = 1 / math.sqrt(2 * math.pi) * math.e ** ((-1/2) * (x**2))
sns.lineplot(x=x, y=y)
```

```python
# 使用 np.random.normal(...) 生成常態分配的隨機變數
# loc=期望值, scale=標準差, size=樣本數
np.random.seed(42)
arr_norm = np.random.normal(loc=10, scale=2, size=500)
pdf_norm = sns.histplot(data=arr_norm, kde=True, bins=30)
```
