```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
sns.set() # activate Seaborn
```

```python
rng = np.random.RandomState(3)
x = 10 * rng.rand(50) # 50 random numbers between 0 and 1
y = 3 + 2 * x + rng.randn(50) # y=3+2x + 常態分配隨機數
```

```python
plt.scatter(x, y) # scatter plot
plt.show()
```

```python
model = LinearRegression(fit_intercept = True)

model.fit(x[:, np.newaxis], y) # 將 x 轉換為 column，建立模型

xfit = np.linspace(0, 10, 1000) # 生成 1000 個 0-10 的等差數列
yfit = model.predict(xfit[:, np.newaxis]) # 以建立好的模型生成對應的預測值

plt.scatter(x, y)
plt.plot(xfit, yfit, color = "red") # 畫出建立好的迴歸直線
plt.show()
```

```python
print("Slope:", round(model.coef_[0], 3))
print("Intercept:", round(model.intercept_, 3))
```

### 多元線性迴歸

```python
rng = np.random.RandomState(1)
X = 10 * rng.rand(10, 3)          # 生成 10*3 的 X
y = 0.5 + np.dot(X, [1.5, -2, 1]) # np.dot()執行矩陣乘法

model = LinearRegression(fit_intercept = True)
model.fit(X, y)

print(round(model.intercept_, 2))
print(model.coef_)
```
